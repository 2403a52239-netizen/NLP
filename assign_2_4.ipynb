tokenization

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
medical_text = """
Hypertension is a chronic medical condition in which the blood pressure in the arteries is persistently elevated.
Diabetes mellitus is a group of metabolic diseases characterized by high blood sugar levels over a prolonged period.
Patients with asthma often experience wheezing, coughing, and shortness of breath due to airway inflammation.
"""
words_t=word_tokenize(medical_text)
print(words_t)
```

```python
from nltk.tokenize import sent_tokenize
sent_t=sent_tokenize(medical_text)
print(sent_t)
```

stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
stemmer = PorterStemmer()
words_t= word_tokenize(medical_text)
stemmed_words = [stemmer.stem(word) for word in words_t]
print(stemmed_words)
```

lemmatization

```python
nltk.download('omw-1.4')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words_t= word_tokenize(medical_text)
for word in words_t:
    print(word,"--->",lemmatizer.lemmatize(word))
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
words_in_quote = word_tokenize(medical_text)
words_in_quote
```
