```python
!pip install nltk spacy
!python -m spacy download en_core_web_sm

import nltk
import spacy
from nltk.tokenize import TreebankWordTokenizer

nlp = spacy.load("en_core_web_sm")

text = """
Artificial intelligence is transforming the world by improving healthcare,
education, and communication. Researchers continue to explore new ways
AI can assist humans in daily life.
"""

text_lower = text.lower()

tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(text_lower)
nltk_word_count = len(tokens)

doc = nlp(text_lower)
spacy_word_count = len([token for token in doc if token.is_alpha])

print(text_lower)
print("NLTK Word Count:", nltk_word_count)
print("spaCy Word Count:", spacy_word_count)
```
